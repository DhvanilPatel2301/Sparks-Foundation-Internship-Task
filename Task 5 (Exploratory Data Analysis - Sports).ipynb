### Plotting Bar Graph For Number Of Matches Played In IPL

```python
plt.figure(figsize = (18,10))
sns.countplot(x='season', data=match_data, palette="cool")
plt.title("Number of Matches played in each IPL season",fontsize=20)
plt.xlabel("season",fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.show()
```

#### Conclusion :
Thus, Here we can observe that <b> Highest Matches played in Year 2013 followed by 2012 and 2011

### Plotting Number Of Matches Won By Team

```python
plt.figure(figsize = (18,10))
sns.countplot(x='winner',data=match_data, palette='winter')
plt.title("Numbers of matches won by team ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Teams",fontsize=15)
plt.ylabel("No of wins",fontsize=15)
plt.show()
```

#### Conclusion 
Here From this graph we can obsereve that <b> Mumbai Indians Won Highest match followed by Chennai Super Kings

### Now we plot pie chart to see percentage of match won by batting and Bowlling decision

```python
Win=match_data.win_by.value_counts()
labels=array(Win.index)
sizes = Win.values
colors = ['#00C957', '#B22222']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal')
plt.show()
```

#### Conclusion :
Thus , <b> Chance of winning is Higher for Decision Bowl First
